```python
#creating image with RGB
from PIL import Image
import random
width = 640
height = 480

img = Image.new('RGB',(width, height))
pix = img.load()


for i in range(width):
    for j in range(height):
        
        r = random.randrange(0,255)
        g = random.randrange(0,255)
        b = random.randrange(0,255)
        
        pix[i,j] = (r,g,b)
        
        
img.save('NewImage.jpg')
img.show()
```

```python
from PIL import Image, ImageOps

# Open the image
img = Image.open("Luffy.jpg")
img.show()

# Convert the image to grayscale
img1 = img.convert("L")
img1.save('LuffyGraylev.jpg')
img1.show()

# Get image dimensions
width, height = img1.size

# Initialize the list to store bit plane images
bit_planes = []
for i in range(8):
    bit_planes.append(Image.new('L', (width, height)))

# Iterate through each pixel and separate the bit planes
for y in range(height):
    for x in range(width):
        pixel_value = img1.getpixel((x, y))
        for bit in range(8):
            bit_value = (pixel_value >> bit) & 1
            if bit_value == 1:
                bit_planes[bit].putpixel((x, y), 255)
            else:
                bit_planes[bit].putpixel((x, y), 0)

# Display the bit planes
bit_planes[0].show(title="Bit plane 0")
bit_planes[1].show(title="Bit plane 1")
bit_planes[2].show(title="Bit plane 2")
bit_planes[3].show(title="Bit plane 3")
bit_planes[4].show(title="Bit plane 4")
bit_planes[5].show(title="Bit plane 5")
bit_planes[6].show(title="Bit plane 6")
bit_planes[7].show(title="Bit plane 7")
```

```python
from PIL import Image, ImageFilter
img = Image.open("Luffy.jpg")
img.show()

img1 = img.convert("L")
img1.save("ImageLuffy10.jpg")
img1.show()

kernel_Orig = [
    [0,0,0],
    [0,2,-1],
    [0,-1,0]
]
kernel =[list(reversed(row)) for row in reversed (kernel_Orig)]
    
for row in kernel:
    print(row)
    
flattened_kernel = [element for row in kernel for element in row ]
print(flattened_kernel)

img2 = img1.filter(ImageFilter.Kernel((3,3), flattened_kernel,1,0))
img2.save("EdgeOrdinalLuffy.jpg")
img2.show()
```
